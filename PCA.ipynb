```python
import pandas as pd
import numpy as np
from numpy import array
from numpy import mean
from numpy import cov
from numpy.linalg import eig
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
```

```python
#here A is a matrix
a = array([[7,9], [9,5], [6,0]])
print(a)
plt.plot([row[0] for row in a],[row[1] for row in a])
```

```python

m = mean(a.T,axis = 1)
print(m)
```

```python
c = a - m
print(c)
```

```python
v = cov(c.T)
print(v)
```

```python
values, vectors=eig(v)
print(vectors)
print(values)
print("h")
p = vectors.T.dot(c.T)
print(p.T)
print("h")
print(p.T[0])
print("h")
print(p.T[1])
print("h")
plt.plot(p.T[:,0],p.T[:,1])
plt.show()
```
